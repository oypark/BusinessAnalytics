## 1. 소셜 네트워크 분석의 개요

### 1) 네트워크 이론의 개요

#### 기본 용어
  - 네트워크: 각각의 객체들이 상호 연결된 구조
  - 구성요소: node(=vertex)와 link(=edge)
  - node: 고유한 속정을 가진 행위자(점)
  - link: 노드들 간의 관계(연결고리, 선)
  - ex) 인터넷 web, 교우관계, 지하철 노선도, 지도 도로망, 항공 노선도, 가계도, 소셜미디어, 장바구니 영수증, 단톡방 등등

#### 네트워크 이론의 개요
  - 네트워크 이론은 수학의 **그래프 이론(graph theory)** 에서 비롯된 이론
    - 그래프 이론은 점(노드)와 그 점을 잇는 선(링크)으로 이루어진 그래프를 수학적으로 연구하는 이론, 수학과 전산학 분야에서 널리 연구됨
  - 월드와이드웹(www), 인터넷, 단백질 상호작용 네트워크, 사회 네트워크 등이 **복잡계 네트워크**를 이룬다는 사실이 알려진 이후로 물리학, 생물학, 셩제학, 사회학 등 여러분야에 복잡계 네트워크 이론이 활발이 적용되고 있음
    - 복잡계 네트워크는 엄청 복잡하게 이루어진 네트워크 안에 일정한 성질을 가지고 있다는 특징이 있음

#### (5) 그라노베터의 '약한 연결(weak ties)의 강점'
  * 미국 스탠포드 대학교 그라노베터 교수는 1973년 자신의 논문 '약한 연결의 강점'에서 **노동자들이 자신의 사회적 네트워크를 활용하여 직업을 어떻게 얻는지**에 대해서 조사 발표하였음
  * 직업을 얻게 된 대부분의 사람들은 자신의 친구나 친척처럼 가까운 사람이 아닌, 그냥 **가볍게 알고 지내는 지인을 통해서 얻는 경우가 훨씬 더 많았음**
  * 기존 상식과 달리, **약한 사회적 관계의 유용성**에 대한 새로운 시사점을 제시. 강한 연결 네트워크들이 클러스터를 이루고, 이러한 **클러스터들은 서로 약한 연결로 연결되는 구조로 사회 네트워크가 구성됨**을 보여줌.
  * 아래 그림 (a)에서 왼쪽과 오른쪽에 강한 연결 네트워크들이 클러스터를 이루고 있음. 이러한 클러스터들을 연결하는 것은 A-B, E-I의 약한 연결임. **인맥관리가 중요하며, 얇고 넓게 많은 사람들을 아는 것이 매우 중요하다는 점을 시사함**

#### (6) 프리만의 '중심성 지표' 개발
  * 프리드만은 **네트워크 상에서 각 노드의 중심성을 측정하는 세 가지 '중심성 지표'의 개념**을 소개하는 논문을 발표함
  * **연결정도 중심성(Degree centrality)**
  * **근접 중심성(Closeness centrality)**
  * **매개 중심성(Betweenness centrality)**: 매개 중심성은 *한 노드가 다른 노드들간의 네트워크 관계에 있어서 중개자, 혹은 매개자 역할을 얼마나 수행하는지* 측정하는 지표
  * 매개자(플랫폼) 기업은 새로운 시장을 창출하기 때문에 항상 성공할 수 밖에 없음 - ex) 배달의 민족, 지그재그, *App store(Apple)*
> mp3 플레이어가 있던 시절, 사람들은 CD 플레이어의 가벼운 버전으로만 mp3 플레이어를 사용하고 있었고, CD 1장 이상 용량을 가질 수 있다는 생각을 전혀 하지 못함. 그 때 스티브 잡스가 애플에 복귀해 **용량을 훨씬 키운 mp3 플레이어인 ipod을 만들어 출시**하게 되었고, 많은 음악을 확보하기 위해 **아이튠즈를 개발하여 음원제작사들에게 유료로음악을 듣게 해준다는 조건하에 사람들을 ipod 공간으로 모으게 됨**. 어떤 사람이 'jail break'를 통해 다른 어플들을 ipod에서 사용할 수 있게 만들었는데, 스티브 잡스는 이에 아이디어를 얻어 **App Store를 만들었고 개발자들과 소비자를 연결하는 플랫폼으로 작용**할 수 있게 되었음. 이후 iphone의 성공에도 많은 영향을 미침.

#### (7) 버트의 '구조적 공백'
  * 사회 네트워크 구조 내에서 상대적으로 중요한 위치와 역할을 하는 노드들을 설명하기 위해 '구조적 공백' 이론 제시
  * **구조적 공백의 위치에 있는 노드**는 다른 노드들과 중복적인 관계를 가지고 있지 않기 때문에 **노드와 노드를 연결시켜주는 중요한 역할**을 함
  * 그라노베티의 '약한 연결(weak ties)' 개념을 더 구체화 한 것
  * **실제 조직내에서** 형성된 개인의 인적 네트워크 구조를 보았을 때, **구조적 공백이 많은 사람이 적은 경우보다 더 빨리 승진**한다는 것을 밝힘.
  <img src="2-7.png">
  * 이와 같은 연구 결과는 **구조적 공백이 네트워크 내에서의 정보 흐름과 영향력에 있어 상대적으로 중요한 역할을 한다는 것을 증명**
  * 아래 그림은 구조적 공백의 관점에서 네트워크를 최적화하기 전과 후를 비교한 것.
  <img src="2-8.png" width="80%">
  * 전과 후 모두 네트워크 중심에 다섯 개의 링크가 존재하지만, 구조에서는 큰 차이가 있음. **최적화 전의 경우 네크워크가 두 개의 소그룹에만 연결**되어 있지만, **최적화 후에는 다섯 개의 링크 모두 서로 다른 다섯 개의 소그룹에 각각 연결**되어 있음
  * 즉 **후자가 더욱 많은 구조적 공백**을 가지게 되며 **효율적인 네트워크 구조로 최적화** 되어있다고 볼 수 있음
> 나는 구조적 공백을 많이 가지고 있는 사람인지? 매개적인 역할을 많이 하는지 생각해볼 필요가 있음. 새로운 단체의 사람들과 친해지는 기회를 많이 가져야함. 약한 관계(weak ties)를 많이 만들고 구조적 공백을 많이 유지하면 새로운 기회를 가질 수 있음.

#### (8) 왓츠, 스트로가츠의 '좁은 세상 네트워크'
  * 전력망, 선충의 신경망, 영화배우 간 연결망의 **연결경로 특성을 분석 하여** 정규 네트워크(Regular network)와 무작위 네트워크(Random network)의 **중간에 존재하는 좁은 세상 네트워크(small world network)를 제시**
  * 정규 네트워크: 인접한 노드와 일정한 규칙에 따라 일정한 숫자로만 연결되어 있음(randomness, 즉 p=0)
  * 무작위 네트워크: 무작위로 연결되어 있음(p=1)
  * **노드들 간에 지름길이 존재하며 이것을 통해 전체 네트워크가 긴밀히 연결**됨. 이후 **무척도 네트워크(scale-free network)가 밝혀지게 되는 계기를 제공**

#### (9) 바라바시, 알버트, 정하웅의 '무척도 네트워크'
  * **하이퍼 링크로 연결된 인터넷의 물리적 연결 구조**에 대한 논문에서 **인터넷 연결 수 분포가 정규 분포를 따르지 않고 멱함수법칙(Power law)을 따르는 것**을 밝혔으며, **이러한 네트워크 구조를** '무작위 네트워크(random network)'와 구별하여 **'무척도 네트워크(scale-free network)'라고 명명함**
  * 무작위 네트워크는 링크의 연결패턴이 평균값을 중심으로 정규분포를 이루는 반면, **무척도 네트워크에서는 소수 링크만을 가진 수많은 노드들과 수많은 링크를 가진 소수 노드들로 양분됨.**

  * **멱함수 분포는** 네트워크 형성에 있어 **선호적 연결이라는 독특한 현상에 기인.** 즉 노드 간에 연결이 이루어 질 때, 아래의 그림 (b)와 같이 특정 노드에 대한 선호적 연결 현상이 발생.
  <img src="2-11.png" width="80%">
  * 이러한 현상이 반복되면 **아주 많은 링크를 가진 소수 노드들이** 형성되고, 이 소수 노드들은 **전체 네트워크에서 허브 등의 중요한 역할 수행.** 위의 항공노선 네트워크 지도를 보면 무척도 네트워크의 특성에 따라 몇몇 소수의 허브공항을 중심으로 선호적 연결을 보이게 됨.
  
>* 2단계 이론(유통이론): 몇몇의 중요한 사람(허브, 인플루언서)에게 정보 등을 전파하는 것이 전체 사람들에게 전파하는 것 보다 효과가 좋음. 즉, 네트워크 상에서 허브를 만들거나 찾는 것이 매우 중요함.

>* 임계치 전까지는 random으로 전파하지만 일정 시점 이후로는 몇몇 node를 중점적으로 네트워크가 형성되고, 나중에는 저절로 굴러가게 됨.

## 2. 네트워크 이론 기초

### 1) 그래프의 기본 개념
  
  * 네트워크를 표현하는 '네트워크 모델링' 과정에서 중요하게 사용되는 것이 그래프, 행렬
  * 그래프의 기본 개념
  >** 그래프 이론(Graph theory): 객체들 간의 네트워크 관계 모델을 수학적 그래프로 나타내는 연구 분야. 그래프는 노드(=node=vertex)로 구성, 노드들은 링크(=link=edge)로 연결되어 있음.
  >** 무방향(Undirected) 네트워크: 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크 등
  >** 방향(Directed) 네트워크: 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등
  
  <img src="2-12.png" width="70%">
  
  * 네트워크 그래프에서 노드간 관계의 유무 뿐만 아니라 관계의 정도(강도)를 함께 표시할 수 있음
  * 관계의 유무만 표현한 경우를 '이진 네트워크(binary network)' 그래프, 관계의 정도를 함께 표시한 경우를 '가중 네트워크(valued network)' 그래프라 함
  * 관계의 정도는 해당 링크 옆에 가중치(숫자)를 직접 표시하거나, 가중치의 정도에 따라 링크 선의 굵기를 다르게 표시하는 방법이 있음
  <img src="2-13.png" width="80%">

### 2) 그래프의 표현 방법

  * 인접행렬(Adfacency matrix), 노드 리스트(Node list), 엣지 리스트(Edge list) 방식이 있음
  * 인접행렬, 인접리스트를 코드로 구현하는 방법
  <https://blog.naver.com/PostView.nhn?blogId=phj8498&logNo=221348733459>
  
#### (1) 인접행렬

  * **인접한 노드들 간의 연결관계를 행렬(Matrix)로 표현.** 네트워크 상에서 **노드의 개수가 n인 경우, 인접 행렬은 nxn의 정방행렬**로 표현
  * 행, 열의 이름은 각 노드를 나타내며, 행과 열이 교차하는 지점의 값은 **노드들간의 관계의 유무를 1, 0**으로 나타냄. **행은 출발** 노드, **열은 도착** 노드. 아래는 **방향 네트워크**를 인접행렬로 나타낸 것.
  <img src="2-14.png" width="50%">
  * **무방향 네트워크의 경우**, 모든 관계가 쌍방향이기 때문에 **인접행렬은 대칭이 됨.**
  <img src="2-15.png" width="70%">

#### (2) 노드 리스트(Node list)

  * 인접한 링크가 있는 노드만을 표기하는 방법, 노드들의 목록
  * 그림의 위쪽 그래프는 무방향 네트워크, 아래쪽 그래프는 방향 네트워크를 그린 것
  <img src="2-16.png" width="50%">
  * 무방향 그래프의 표 왼쪽 열은 그래프에 존재하는 노드들의 목록, 표 안쪽은 그 노드에 연결된 노드들의 목록을 나타냄. 전체 행의 개수 = 전체 노드의 개수. 노드에 연결된 링크의 개수가 많을 수록 열의 개수도 증가.
  * 방향 그래프의 경우 indoming, outgoing을 구분해줌.
